# let's check if user enter number that exist with our database 

```python
def no_exisist(self, phone):
        phone = self.no.get().strip()

        vals = (phone, )
        select_query = "SELECT * FROM `studeny_info` WHERE `PHONE` =%s"
        self.c.execute(select_query, vals)
        user = self.c.fetchone()
        if user is not None:
            return True
        else:
            return False
```

### The code defines a method named "no_exisist" that takes a phone number attribute from the object and checks if it exists in the 'student_info' table of a MySQL database. The method executes a SELECT query on the table to search for the phone number, using the cursor object created earlier. The results of the query are stored in the "user" variable using the "fetchone" method. If the phone number is found in the table, "user" will contain a non-empty tuple and the method will return True. If the phone number is not found, "user" will be None and the method will return False. The phone number is retrieved from the "no" attribute of the object and stripped of any leading or trailing whitespace. Overall, this method is used to validate whether a phone number already exists in a MySQL database table.

## the main function is register 

```python
def register(self):

        # self.id = self.id.get()
        id = self.id.get().strip()  # .strip() mean : remove space from text
        Name = self.name.get().strip()
        phone = self.no.get().strip()
        Email = self.email.get().strip()
        Pass1 = self.pass1.get().strip()
        pass2 = self.pass2.get().strip()
        Level = self.level.get().strip()
        Faculty = self.fc.get().strip()

        if len(id) > 0 and len(Name) > 0 and len(phone) > 0 and len(Email) > 0 and len(Pass1) > 0 and len(Level) > 0 and len(Faculty) > 0:
            if self.no_exisist(phone) == False:
                if Pass1 == pass2:
                    vals = (id, Name, phone, Email, Pass1, Level, Faculty)
                    insert_query = "INSERT INTO `studeny_info`(`id`,`NAME`, `PHONE`, `EMAIL`, `PASSWORD`, `LEVEL`, `FACULTY`) VALUES (%s,%s,%s,%s,%s,%s,%s)"
                    self.c.execute(insert_query, vals)
                    self.connection.commit()
                    messagebox.showinfo(
                        'register', 'your account has been registered successfully ' + ' mr/ms : ' + Name)
                    root.destroy()
                    import main
                else:
                    messagebox.showerror(
                        'password', 'your password is not same')
            else:
                messagebox.showwarning(
                    'phone', 'phone number already exist try another')
        else:
            messagebox.showerror(
                'register', 'your account has not been registered  ')
```
